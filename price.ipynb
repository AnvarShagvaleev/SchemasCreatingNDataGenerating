```python
data = pd.read_excel('NewData.xlsx', sheet_name='sr.2', index_col='id')
names = data['name']
names.iloc[44]
```

```python
# .send_keys('names.iloc[44]')
prcs = []
for i, name in enumerate(names):
    logger.info(f'{i} / {len(names)}')
    search = driver.find_elements("xpath", '//input[@class="gLFyf"]')[0]
    button = driver.find_elements("xpath", '//button[@jsname="Tg7LZd"]')[0]
    search.clear()
    search.send_keys(name)
    button.click()
    time.sleep(3)
    try:
        price = driver.find_elements("xpath", '(//div[@class="XrAfOe"])[1]')[0]
        prc = price.text
    except:
        prc = None
    prcs.append(prc)
```

```python
nprcs = []
for i, p in enumerate(prcs):
    logger.info(f'{i+1} / {len(names)}')
    if p == None:
        nprcs.append(p)
    else:
        nprcs.append(p.split(',')[0].replace(' ', ''))
    
```

```python
dfprice = pd.DataFrame(nprcs, columns=['price'])
```

```python
data['price'] = dfprice
```
